```python
df = pd.read_csv("metereo.csv", encoding="utf-8", skiprows=2)
df.head()
```

```python
victimas_df = pd.read_csv('victimas_siniestros.csv', encoding='latin-1', engine='python', sep=';')
victimas_df.head()
```

```python
siniestros_df = pd.read_csv('siniestros_viales.csv', encoding='latin-1', engine='python', sep=';')
siniestros_df.head()
```

```python
siniestros_victimas_df = pd.merge(siniestros_df, victimas_df, on='id_hecho', how='inner')

print("Head del dataframe:")
display(siniestros_victimas_df.head())

print("\nColumnas del dataframe:")
print(siniestros_victimas_df.columns)
```

```python
siniestros_victimas_df['hora'] = siniestros_victimas_df['hora'].str.zfill(8)
siniestros_victimas_df['timestamp'] = siniestros_victimas_df['fecha_x'].astype(str) + ' ' + siniestros_victimas_df['hora']
siniestros_victimas_df['timestamp'] = pd.to_datetime(siniestros_victimas_df['timestamp'], errors='coerce')

df['time'] = pd.to_datetime(df['time'], errors='coerce')

print("Data types de siniestros_victimas_df:")
print(siniestros_victimas_df[['fecha_x', 'hora', 'timestamp']].dtypes)
print("\nData types de df:")
print(df['time'].dtypes)
```

```python
siniestros_victimas_df.dropna(subset=['timestamp'], inplace=True)

df.dropna(subset=['time'], inplace=True)

siniestros_victimas_meteorologia_df = pd.merge_asof(siniestros_victimas_df.sort_values('timestamp'),
                                                    df.sort_values('time'),
                                                    left_on='timestamp',
                                                    right_on='time',
                                                    direction='nearest')

print("Head del dataframe fusionado:")
display(siniestros_victimas_meteorologia_df.head())

print("\nColumnas del dataframe fusionado:")
print(siniestros_victimas_meteorologia_df.columns)
```

```python
print("Head del dataframe fusionado:")
display(siniestros_victimas_meteorologia_df.head())

print("\nColumnas del dataframe fusionado:")
print(siniestros_victimas_meteorologia_df.columns)
```

```python
siniestros_victimas_meteorologia_df['timestamp'] = pd.to_datetime(siniestros_victimas_meteorologia_df['timestamp'])

date_to_verify = '2019-01-01'
filtered_df = siniestros_victimas_meteorologia_df[siniestros_victimas_meteorologia_df['timestamp'].dt.date == pd.to_datetime(date_to_verify).date()]

print(f"Data for {date_to_verify}:")
display(filtered_df)
```
